```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_execution_times(filename, p):
    data = np.loadtxt(f'../execution_times/{filename}.csv', delimiter=',')
    x = data[:, 0]  # First column (x-values)
    y = data[:, 1]  # Second column (y-values)
    # Create the plot
    plt.plot(x, y, marker='o')  # You can customize the plot with markers or line styles

    # Add labels and a title
    plt.xlabel('Text Length (number of chars)')
    plt.ylabel('Execution time (s)')
    plt.title(f"BWT Pattern Matching - pattern '{p}'")
    plt.show()
```

```python
plot_execution_times("execution_times_battle", "battle")
```

```python
plot_execution_times("execution_times_pt", "pt")
```

```python
plot_execution_times("execution_times_patroklos_big", "patroklos")
```
